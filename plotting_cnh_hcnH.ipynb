```python
# importing the required module
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(10,8)) 

x = [3, 5, 7, 9, 11]#carbon chain lengths

#include c=5 bc h ow else do i connect the points lol
kelvx=[5, 7,9]
kelvy = [np.log10(1.3e12),10.11, 9.88]
myx = [5, 7,9,11]
myy = [np.log10(1.3e12),10.983356903926, np.log10(9.234e11), (8.9248)]






# CnH, n = odd numbers
y = [np.log10(1.82e12),np.log10(1.3e12), np.log10(4e11)] #measured/upper limt
plt.plot(x[2:3], y[2:3], 'gs', markersize=12, label = 'Upper Limit') #plots the upper limit from i think mccarthy paper






# HCnN, n = odd
y2 = [np.log10(1.8e14),np.log10(6.69e13), np.log10(3.65e13), np.log10(2.16e13), np.log10(1.04e12)]


#plt.plot(x, y,'go-', linewidth=2, markersize=12, label='CnH')#plotting CnH

plt.plot(kelvx, kelvy, 'yo--', label='CnH (KL-ML predicted CD)')

plt.plot(x[:3], y[:4], 'go-' ,label='CnH (Measured Column Density)')
plt.plot(myx, myy, 'b^--', label='CnH (ML-Predicted Column Density)', markersize=10)
plt.plot(x,y2, 'ro-', linewidth=2,  label='HCnN') #plotting HCnN




# naming the x axis
plt.xlabel('number of carbons')
# naming the y axis
plt.ylabel('log column densities')

plt.legend()

# giving a title to my graph
plt.title('Carbon chain lengths vs CD')

# function to show the plot
plt.show()
```
